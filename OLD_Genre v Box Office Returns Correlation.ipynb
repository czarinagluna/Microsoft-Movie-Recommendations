### Merge data sets

Our goal is to find the most successful movie genre in terms of box office returns. In order to find this, we must first identify th emovies that fit a 1:1.25 ratio of budget to box office return and then identify the genres which are most successful given that ratio. 

We decided to merge the data sets which contained all the relevant information that would allow us to make this correlation.

```python
# Merging on tconst
x = title_basic.merge(tittle_ratings, on="tconst")
```

```python
#Renaming primary_title to movie for next merge
x = x.rename(columns={"primary_title":"movie"})
title_crew = title_crew.rename(columns={"directors":"nconst"})
```

```python
name_director = title_crew.merge(name_basic, on="nconst")
```

```python
imdb_df = x.merge(name_director, on="tconst")
```

```python
full_df = movie_budgets.merge(imdb_df, on="movie")
full_df
```

```python
df = imdb_df.merge(movie_budgets, on="movie")
df
```

### Introducing the succes ratio

### Turning strings into floats

Before merging, we turned the string values for worldwide_gross, domestic_gross, and production_budget into floats so we can work with them later on

```python
df["worldwide_gross"] = df["worldwide_gross"].str.replace('$', '')
df["domestic_gross"] = df["domestic_gross"].str.replace('$', '')
df["production_budget"] = df["production_budget"].str.replace('$', '')
df["worldwide_gross"] = df["worldwide_gross"].str.replace(',', '')
df["domestic_gross"] = df["domestic_gross"].str.replace(',', '')
df["production_budget"] = df["production_budget"].str.replace(',', '')
df["production_budget"]= df["production_budget"].astype(float)
df["domestic_gross"]= df["domestic_gross"].astype(float)
df["worldwide_gross"]= df["worldwide_gross"].astype(float)
```

```python
df
```

```python
df['s_movie_ratio'] = df['worldwide_gross'] / df['production_budget']
df
```

```python
s_movies_df = df.loc[df['s_movie_ratio'] >= 1.25]
s_movies_df
```

```python
s_movies_df["genres"].replace('NaN', np.nan).dropna(inplace = True)
s_movies_df
```
