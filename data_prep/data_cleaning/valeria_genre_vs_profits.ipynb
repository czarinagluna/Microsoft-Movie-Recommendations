### Cleaning the data
Our goal is to find the most successful movie genres in terms of box office returns. In order to find this, we must first identify the movies that returned a profit and then identify the genres which are most successful given that ratio.

First, I merged the data sets in order to have all the informaton needed for this argument.

```python
df = tn_gross.merge(tn_production, on="Unnamed: 0")
df
```

#### Turning strings into floats
After merging, I turned the string values for worldwide_gross, domestic_gross, and production_budget into floats so we can work with them later on.

```python
df["worldwide_gross"] = df["worldwide_gross"].str.replace('$', '')
df["domestic_gross"] = df["domestic_gross"].str.replace('$', '')
df["production_budget"] = df["production_budget"].str.replace('$', '')
df["worldwide_gross"] = df["worldwide_gross"].str.replace(',', '')
df["domestic_gross"] = df["domestic_gross"].str.replace(',', '')
df["production_budget"] = df["production_budget"].str.replace(',', '')
df["production_budget"]= df["production_budget"].astype(float)
df["domestic_gross"]= df["domestic_gross"].astype(float)
df["worldwide_gross"]= df["worldwide_gross"].astype(float)
```

I added a column to denote the retun on investment for each movie.

```python
df['earnings_ratio'] = df['worldwide_gross'] / df['production_budget']
df
```

#### Cleaning Genres

```python
genres_df = df
genres_df = genres_df[genres_df.genres != "None"]
genres_df.head(2)
```

### Using Median profits instead of sum of profits

```python
median_df = genres_df.groupby("genres", as_index=False).median()
median_df.loc[:,["genres","worldwide_gross"]]
```

```python
plt.figure(figsize=(10, 10))
plt.bar(
    x=median_df["genres"], 
    height=median_df["worldwide_gross"])
plt.xticks(rotation=45)
plt.ticklabel_format(axis="y", style="plain")
;
```

```python
mean_df = genres_df.groupby("genres", as_index=False).mean()
mean_df.loc[:,["genres","worldwide_gross"]]

plt.figure(figsize=(10, 10))
plt.bar(
    x=mean_df["genres"], 
    height=mean_df["worldwide_gross"])
plt.xticks(rotation=45)
plt.ticklabel_format(axis="y", style="plain")
;
```
